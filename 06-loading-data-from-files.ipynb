```python
import matplotlib.pyplot as plt
import numpy as np
import csv

x = []
y = []

with open('example.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    print(plots)
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
plt.plot(x,y, label='the normal way')
plt.figure(1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot')
plt.legend()
plt.show()

x,y = np.loadtxt('example.csv', delimiter=',', unpack=True)



plt.figure(2)
plt.plot(x,y, label='the numpy way')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot')
plt.legend()
plt.show()



```
